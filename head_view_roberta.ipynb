```python
from bertviz.pytorch_transformers_attn import RobertaModel, RobertaTokenizer
from bertviz.head_view import show
```

```python
%%javascript
require.config({
  paths: {
      d3: '//cdnjs.cloudflare.com/ajax/libs/d3/3.4.8/d3.min',
      jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
  }
});
```

```python
model_type = 'roberta'
model_version = 'roberta-base'
# model_version = 'roberta-large'
# model_version = 'roberta-large-mnli'
model = RobertaModel.from_pretrained(model_version)
tokenizer = RobertaTokenizer.from_pretrained(model_version)
sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
show(model, model_type, tokenizer, sentence_a, sentence_b)
```
